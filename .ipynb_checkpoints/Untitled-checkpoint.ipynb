```python
import sys
import numpy as np

FILENAME = "Data/2019WAS.EVN"

f = open(FILENAME, 'r')
homePitcher = None
awayPitcher = None
gameCtr = 0
atBats = []

for line in f:
    line = line.strip().split(',')

    #newGame
    if line[0] == 'id':
        gameCtr += 1

    #starting pitcher
    if line[0] == 'start' and line[-1] == '1':
        if line[-3] == '0':
            awayPitcher = line[1]
        elif line[-3] == '1':
            homePitcher = line[1]

    #Pitching change
    if line[0] == 'sub' and line[5] == '1':
        if line[3] == '1':
            homePitcher = line[1]
        elif line[3] == '0':
            awayPitcher = line[1]

    #Batter
    # if line[2] == 0
    if line[0] == 'play':
        batter = line[3]
        if line[2] == "0":
            pitcher = homePitcher
        elif line[2] == "1":
            pitcher = awayPitcher

    #Hit or not
        if line[-1][0] in ["S", 'D', 'T', 'H'] or line[-1][0:2] == 'HR':
            event = "Hit"
        elif line[-1][0] in ["1", "2", "3", "4", "5", "6", "7", "8", "9"]:
            event = 'Out'
        else:
            event = "skip"

        #print(f"Matchup: {batter} vs {pitcher} -- result: {event}")
        if event != "skip":
            atBats.append((batter, pitcher, event))
            
    
```

```python
atBats = pd.DataFrame(atBats, columns=['batter', 'pitcher', 'event'])
```
